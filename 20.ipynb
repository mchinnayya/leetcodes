```python
# Easy method to understand
def isValid(s):
    stack = []
    match = {')': '(', '}': '{', ']': '['}

    for char in s:
        # 1️⃣ Opening bracket → push
        if char in '({[':
            stack.append(char)
        
        # 2️⃣ Closing bracket → check match
        elif char in match:
            if len(stack) == 0:        # no opening
                return False
            if stack[-1] != match[char]:  # mismatch
                return False
            stack.pop()  # matched pair → remove
        
        else:
            return False  # invalid character
    
    # 3️⃣ After all characters
    return len(stack) == 0
```

```python
def isValid(s):
    stack = []
    map = {')': '(', '}': '{', ']': '['}

    for char in s:
        if char in map.values():
            stack.append(char)
        elif char in map:
            if len(stack) == 0 or stack[-1] != map[char]:
                return False
            stack.pop()
        else:
            return False
    return len(stack) == 0  # also no 'not'
```

```python
def isValid(s):
    stack = []
    map = {')': '(', '}': '{', ']': '['}
    
    for char in s:
        if char in map.values():
            stack.append(char)
        elif char in map:
            if not stack or stack[-1] != map[char]:
                return False
            stack.pop()
        else:
            return False  
    return not stack

print(isValid("()[]{}"))
```
