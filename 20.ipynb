```python
def isValid(s):
    stack = []
    map = {')': '(', '}': '{', ']': '['}

    for char in s:
        if char in map.values():
            stack.append(char)
        elif char in map:
            if len(stack) == 0 or stack[-1] != map[char]:
                return False
            stack.pop()
        else:
            return False
    return len(stack) == 0  # also no 'not'
```

```python
def isValid(s):
    stack = []
    map = {')': '(', '}': '{', ']': '['}
    
    for char in s:
        if char in map.values():
            stack.append(char)
        elif char in map:
            if not stack or stack[-1] != map[char]:
                return False
            stack.pop()
        else:
            return False  
    return not stack

print(isValid("()[]{}"))
```
